```python
import torch
import cv2
import numpy as np
import os
from PIL import Image
from IPython.display import display
```

```python
# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Define classes
classes = ['car', 'hiace']

# Define threshold for object confidence
confidence_threshold = 0.5

# Loop through images in folder
image_folder = 'D:\EducationalStuff\MastersDS\HisThesis\AllDataSets\CustomizedData\CarImages'
for filename in os.listdir(image_folder):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # Load image
        img = Image.open(os.path.join(image_folder, filename))
        display(img)
        img = np.array(img)

        # Perform object detection
        results = model(img)

        # Loop through detected objects
        for result in results.xyxy[0]:
            # Extract class and confidence score
            class_id = int(result[5])
            confidence = result[4]

            # Filter objects by confidence score
            if confidence > confidence_threshold:
                # Extract object coordinates
                x1, y1, x2, y2 = map(int, result[:4])

                # Crop object from image
                object_img = img[y1:y2, x1:x2]
                display(object_img)

                # Save object image
                object_filename = f'{filename[:-4]}_{classes[class_id]}.jpg'
                object_path = os.path.join(image_folder, object_filename)
                cv2.imwrite(object_path, object_img)
```
